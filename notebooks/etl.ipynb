# Se procede a desanidar el dataset para generar un dataset por cada columna, posterior a limpiar y unir al dataset correspondiente

```python

# Load the dataset
select_column_movies = pd.read_csv('C:/PI1/data/processed/selected_columns_movies.csv')

# Function to expand a JSON-like column
def expand_json_column(column_data):
    results = []
    for value in column_data.dropna():
        try:
            parsed_data = json.loads(value.replace("'", "\""))  # Convert single quotes to double for valid JSON
            if isinstance(parsed_data, list):
                results.extend(parsed_data)  # Add all elements if it's a list
            elif isinstance(parsed_data, dict):
                results.append(parsed_data)  # Add the single dictionary if it's a dict
        except json.JSONDecodeError:
            continue
    return pd.json_normalize(results) if results else pd.DataFrame()

# Process the columns in the dataset order
processed_files = []

for column in select_column_movies.columns:
    try:
        # Expand the current column
        expanded_df = expand_json_column(select_column_movies[column])
        if not expanded_df.empty:
            # Save the expanded column data to a CSV file
            file_path = f'C:/PI1/data/processed/clear_csv/expanded_{column}.csv'
            expanded_df.to_csv(file_path, index=False)
            processed_files.append(file_path)
            print(f"Processed and saved column '{column}' to: {file_path}")
    except Exception as e:
        print(f"Error processing column '{column}': {e}")

# Output the paths of processed files
print("Processed files:", processed_files)
```
